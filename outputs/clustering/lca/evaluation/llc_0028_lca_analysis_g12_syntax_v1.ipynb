# Evaluation of Latent Class Analysis: Covid > 12 weeks ago

This notebook contains plots that evaluate the performance of latent class analysis (LCA). For each study, LCA has been applied to self-reported symptom data with number of classes ranging from 1 to 5. 

For each study, there is a plot evaluating the LCA. The size of the smallest class, as a % of sample size, is printed for each value of N to ensure the smallest class is > 5% of sample for the chosen N.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.chdir('S:/LLC_0028/data/lca_results')
```

```python
def find_smallest_class(df):
    
    sc = []
    
    for col in df.columns[:-1]:
        
        sc.append(min(df[col].value_counts()))
        
    return np.array(sc)*100/sc[0]
```

```python
def eval_plot(f):
    
    num2cohort = {1: 'bcs70',
                  2:'bib',
                  3: 'mcs',
                  4: 'ncds',
                  5: 'nextstep',
                  6: 'nhsd46',
                  7: 'track19',
                  8: 'twins',
                  9: 'alspac',
                 }
    
    
    dta = pd.read_csv(f, index_col=0)

    class_dta = pd.read_csv(f.split('.')[0] + '_classes.csv', index_col=0)
    
    sc = find_smallest_class(class_dta)
    print('Smallest cluster (% sample size) ',[round(v,2) for v in sc])
    
    fig,ax_list = plt.subplots(1,2, figsize=(8,4))

    ax = ax_list[0]
    ax.plot(dta.Model, dta.BIC, '-.', lw = 2, label='BIC')
    ax.plot(dta.Model, dta.AIC, '-.', lw = 2, label='AIC')
        
    ax.grid()
    ax.set_xlabel('N classes')
    ax.legend(loc='best')

    ax = ax_list[1]
    ax.plot(dta.Model, dta.entropy/dta.entropy.values[0], 'g-.', lw=2, label='Entropy')
    ax.plot(dta.Model, dta.likelihood_ratio/dta.likelihood_ratio.values[0], 'r-.', lw=2, label='LR')

    ax.grid()
    plt.xlabel('N classes')
    plt.legend(loc='best')
 
    f = f.split('.')[0].split('_')
    
    if f[-1] in np.char.mod('%d',np.arange(1,10)):
        
        f[-1] = num2cohort[int(f[-1])]

    fig.suptitle((' ').join(f), fontsize=16)
    plt.tight_layout()

    plt.show()
```

```python
results = pd.DataFrame()
```

```python
cohort2num = {'bcs70':1,
              'bib':2,
              'mcs':3,
              'ncds':4,
              'nextstep':5,
              'nhsd46':6,
              'track19':7,
              'twins':8,
              'alspac':9,
                }
```
