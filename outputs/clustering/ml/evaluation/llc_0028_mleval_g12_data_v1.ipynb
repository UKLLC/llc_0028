# Evaluation of ML clustering algorithms: COVID > 12 weeks ago

This notebook contains plots that evaluate the performance of ML clustering algorithms. For each study, two algorithms (DBSCAN and Agglomerative) have been applied to self-reported symptom data for a range of parameter values. The parameter varied for DBSCAN is 'eps' and the parameter varied for Agglomerative is 'N clusters'.

For each algorithm, and each study, there is a plot displaying the algorithm's performance over a range of parameter values. The evaluation metrics (DB Index and Silhouette Coefficient) are described below. In addition, we've recorded the proportion of study participants not assigned to a cluster (i.e. marked as 'noise' by the algorithm) for each parameter value. Left hand plots show  variation in these evaluation metrics with parameter value. Right hand plots show sizes of clusters, as a fraction of sample size, for each parameter value. 

## Selecting the best parameters 

**Silhouette Score**

$$ S = \frac{b-a}{\max(a,b)} $$

- $a$: mean distance between points within a cluster
- $b$: mean distance between a point and points in the nearest cluster

Bounded between $-1$ and $1$. 

- $s=-1$ indicates incorrect clustering 
- $s=0$ indicates overlapping clusters
- $s=1$ indicates dense, well seperated clusters

**Davies Bouldin Index**

The average similarity between each cluster and its most similar one, averaged over all clusters. 

$$ DB = \frac{1}{k} \sum_{i=1}^{k} \max_{i \neq j} R_{i,j} $$

where

$$ R_{i,j} = \frac{s_i + s_j}{d_{i,j}} $$

- $k$: the number of clusters
- $s_i$: the average distance between each point in cluster $i$ and the cluster centroid
- $d_{i,j}$: the distance between the centroids of cluster $i$ and $j$

Low values indicate a better partitioning of the data 

**Proportion of outliers**

As we are forcing a minimum cluster size, not all datapoints will be assigned to a cluster. Those not assigned are described as outliers.



### Rules

1. maximise S (elbow)
2. minimise DB (elbow)
3. where possible, keep % outliers < 40


#### Edge cases

- If its not clear which parameter is best, ensure $S > 0$ while minimising % outliers

- Where elbows occur for different parameter values in each metric, take the average of elbow parameter values
