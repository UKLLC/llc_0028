## Symptom prevalence 

```python
files = ['llc_0028_symptom_prevalence_nocovid_data_v1_processed.csv',
        'llc_0028_symptom_prevalence_l12_data_v1_processed.csv',
        'llc_0028_symptom_prevalence_g12_data_v1_processed.csv']

categories = ['No history of COVID-19',
              'Recent COVID-19 (< 12 weeks)',
              'Past COVID-19 ($\geq$ 12 weeks)']
```

```python
def _round(x, base=5):
    return base * round(x/base)
```

```python
combined = pd.DataFrame()
for i, fname in enumerate(files):
    df = pd.read_csv(derived_dir + fname, index_col=0)
    df = df.T
    df = df.rename(columns = dict(zip(df.columns.values, 
                            [c.split('_')[0] for c in df.columns.values])))
    df = df.T
    
    for c in df.columns:
        
        new = []
        
        for value in df[c].values:
            
            if value == -1:   
                new.append('<10 (-)')
            else:
                value = _round(value)
                perc = round(value*100 / sum(df[c]),2)
                new.append(f'{int(value)} ({perc})')
            
        df[c] = new
    
    header = pd.DataFrame(index = df.columns)
    header[categories[i]] = ['-' for n in range(len(header))]
    df = pd.concat([header.T, df])
    combined = pd.concat([combined, df])
```

```python
combined
```

```python
combined.T.to_csv(figure_dir + 'symptom_prevalence_table.csv')
```
