```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.chdir('S:\LLC_0028\data')
```

```python
dta = pd.read_csv('./harmonised_all/llc_0028_full_harmonised_data_all_v2.csv', index_col=0)
```

```python
core_symp = ['fever','cough','throat',
            'chest_tight','breath','nose','aches',
             'fatigue', 'diarrhoea','smell_taste',
            'nausea_vomit', 'sneezing',
             'headache','concentrating','memory']
```

```python
dta = dta[core_symp + ['study','covid_status']] #only need symptoms, study and status
```

```python
dta2 = dta.dropna(how='any',subset=core_symp)
```

## Number of people with > n symptoms 

```python
symptom_df = pd.DataFrame()

n_all = []
for n in range(len(core_symp)):
    
    #number of people with >n symtoms
    n_all.append(dta.loc[dta[core_symp].sum(axis=1)>n].shape[0])
    
symptom_df['pooled'] = n_all

for study in dta.study.unique():
    
    n_study = []
    for n in range(len(core_symp)):
    
        #number of people with >n symtoms
        n_study.append(dta.loc[(dta[core_symp].sum(axis=1)>n) 
                               & (dta.study == study)].shape[0])
    symptom_df[study] = n_study
```

```python
symptom_df[symptom_df <10] = '<10'
```

```python
symptom_df.to_csv('./symptom_analysis/llc_0028_nsymptoms_data.csv')
```
