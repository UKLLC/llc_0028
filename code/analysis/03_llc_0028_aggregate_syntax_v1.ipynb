```python
import os
import pandas as pd
```

```python
os.chdir('S:/LLC_0028/data/harmonised_symptom_data/harmonised_symptom/')
```

```python
combined=pd.DataFrame()

for i, file in enumerate(os.listdir()):
    
    if 'v1' in file:
        continue
    if 'v2' in file:
        continue
    
    print(file)
    study = file.split('_')[2][:-2]
    
    df = pd.read_csv(file, index_col=0)
    
    df['study'] = [study for c in range(df.shape[0])]
    
    if 'responsedate' in df.columns:
        
        df = df.drop(['responsedate'],axis=1)
        
    if 'llc_0028_stud_id' in df.columns:
        
        df = df.rename(columns = {'llc_0028_stud_id':'LLC_0028_stud_id'})  
 
    combined = pd.concat([combined, df], ignore_index=True)    
```

```python
harmonised = ['fever','cough','throat','chest_tight',
             'breath', 'nose', 'aches',
             'fatigue', 'diarrhoea', 'smell_taste',
             'nausea_vomit', 'rash', 'sneezing',
             'headache', 'appetite', 'tummy_ache',
             'eyes', 'voice', 'dizzy', 'chest_pain',
             'chills', 'sleep', 'numb', 'heavy', 
             'swelling', 'concentrating', 'memory']
```

```python
combined = combined[['LLC_0028_stud_id', 'study'] + harmonised]
```

```python
combined.shape
```

```python
combined = combined.dropna(how='all',subset=combined.columns[2:])
```

```python
#combine alspac_m and alspac_y
dta = combined.copy()

alsp = dta.loc[dta.study.isin(['alspacm','alspacy'])]

alsp['study'] = ['alspac' for i in range(alsp.shape[0])]

dta = pd.concat([dta, alsp])

dta = dta.loc[~dta.study.isin(['alspacm','alspacy'])]

dta.shape[0] == combined.shape[0]
```

```python
dta.to_csv('../llc_0028_harmsymp_data_v3.csv')
```
