## Process stata results to csv 

```python
result_files = os.listdir('./stata_results/')
```

```python
def process_stata(fname):
    
    df = pd.read_excel('./stata_results/' + fname, header=2, skipfooter=15)
    df = df.rename(columns = {'Unnamed: 1':'OR','Unnamed: 2':'significance'})
    
    se = df.loc[df[df.columns[0]].isna()]['OR'].values[0::3]
    ci = df.loc[df[df.columns[0]].isna()]['OR'].values[1::3]
    p = df.loc[df[df.columns[0]].isna()]['OR'].values[2::3]
    df = df.loc[~df[df.columns[0]].isna()]
    df['SE'] = se
    df['CI'] = ci
    df['p-value'] = p
    df = df.fillna('-')
    df.to_csv('./stata_results/processed/' + fname.split('.')[0] + '.csv', index=False)
    return df
```

```python
for fname in result_files:
    if '.xlsx' in fname:
        if 'complete' in fname:
            process_stata(fname)
```

## Aggregate results 

### No covid 

```python
files = [f for f in os.listdir('./stata_results/processed/') if 'no' in f and 'complete' in f] 
mlr_dta = pd.read_csv('no_covid_mlr_complete_dta.csv', index_col=0)
```

```python
disclosive = []

for var_ in ['llc_sex','llc_ethnic3','age_cat_numeric','functional_limitation_cat']:
    
    for cluster, group in mlr_dta.groupby('lca_cluster'):    
        for value in  group[var_].unique():
            if group.loc[group[var_] == value].shape[0]<10:
                disclosive.append([cluster, var_, value])
```

```python
aggregated = pd.DataFrame()

for fname in files:
    
    df = pd.read_csv('./stata_results/processed/' + fname.split('.')[0] + '.csv')
    cluster = int(fname.split('_')[0][-1])
    var_ = df.columns[0]
    df = df.rename(columns = {var_:'value'})
    df['cluster'] = [cluster for i in range(df.shape[0])]
    df['variable'] = [var_ for i in range(df.shape[0])]
 
    aggregated = pd.concat([aggregated,df])
    
#remove disclosive
for row in disclosive:
    
    aggregated = aggregated.loc[~((aggregated.cluster==row[0]) &
                            (aggregated.variable==row[1]) &
                            (aggregated.value==row[2]))]
    
aggregated.to_csv('./stata_results/processed/aggregated/' + 'llc_0028_no_mlr_complete_results.csv', index=False)
```
