```python
import os
import pandas as pd
```

## Import data 

```python
dta = pd.read_csv('../../data/harmonised_all/llc_0028_full_harmonised_data_all_v2.csv', index_col=0)

core_symp = ['fever','cough','throat',
            'chest_tight','breath','nose','aches',
             'fatigue', 'diarrhoea','smell_taste',
            'nausea_vomit', 'sneezing',
             'headache','concentrating','memory']

dta = dta.dropna(how='any',subset=core_symp)
```

## Create tables 

```python
results = pd.DataFrame(index = dta.study.unique())

n_self, n_ehr, n_self_only, n_ehr_only = [],[],[],[]
sample = []

for study in results.index.values:
    
    df = dta.loc[dta.study==study]
    
    sample.append(df.shape[0])
    
    n_self.append(df.loc[df.self_report_status.isin([1.,2.])].shape[0])
    n_ehr.append(df.loc[df.ehr_status.isin([1.,2.])].shape[0])
    if n_ehr[-1]<10:
        n_ehr[-1]= 0
    n_self_only.append(df.loc[(df.self_report_status.isin([1.,2.])) & ~(df.ehr_status.isin([1.,2.]))].shape[0])
    n_ehr_only.append(df.loc[~(df.self_report_status.isin([1.,2.])) & (df.ehr_status.isin([1.,2.]))].shape[0])
    if n_ehr_only[-1]<10:
        n_ehr_only[-1]=0
    
results['N_self_report'] = n_self
results['sample size'] = sample
results['N_ehr'] = n_ehr
results['N_self_report_only'] = n_self_only
results['N_ehr_only'] = n_ehr_only
```

```python
results.to_csv('../../data/harmonised_all/llc_0028_status_comparison_data_v1.csv')
```
